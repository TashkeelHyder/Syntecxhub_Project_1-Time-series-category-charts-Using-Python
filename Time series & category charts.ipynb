<h2> Monthly line Chart</h2>

- <h3>Extracting month and year from order date</h3>

```python

# Now extract the month and year
df['month_year'] = df['order_date'].dt.to_period('M').astype(str)
```

```python
df['month_year'] 
```

```python
# taking sum of Sales by month and year using Group by
df_trend = df.groupby('month_year')['sales'].sum().reset_index()
```

```python
# setting the Figure size of the Graph
plt.figure(figsize=(16, 5))
plt.plot(df_trend['month_year'], df_trend['sales'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.savefig('month_year.png')
plt.show()
```

- <h4>Creating  a [Quarter_year] Column

```python
# creating Quarter_year coloumn
df['quarter_year'] = df['order_date'].dt.to_period('Q').astype(str)
```

```python
# grouping by Quarter_year column for the  Quarterly Sales data 
df_quarterly = df.groupby('quarter_year')['sales'].sum().reset_index()
```

```python
#  Setting the Figure size
plt.figure(figsize=(12, 6))

plt.plot(df_quarterly['quarter_year'], df_quarterly['sales'], color='#b80045', marker='o', linewidth=2)


plt.title('Quarterly Sales Performance (2011-2014)', fontsize=14, fontweight='bold')
plt.xlabel('Quarter Year', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45) # Rotating for better legibility
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('quarterly_sales_trend.png')

plt.show()
```

<h3> Using bar charts to compare categories and pie
charts for share.</h3>

- <h4>Using the Bar Chart To Compare Categories</h4>

```python
# Grouping sales by Category
category_analysis = df.groupby('category')['sales'].sum().reset_index()
```

```python
# Sorting the Category Sales Data
category_analysis = category_analysis.sort_values(by='sales', ascending=False)
print(category_analysis)
```

```python
# Plotting the Chart
```

```python
import matplotlib.ticker as ticker

# Setting the figure size
plt.figure(figsize=(10, 6))

sns.barplot(data=category_analysis, x='category', y='sales', hue='category', palette='magma', legend=False)

plt.title('Total Sales Comparison by Category', fontsize=14, fontweight='bold')
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))

plt.tight_layout()
plt.savefig('category_comparison_bar.png')

plt.show()
```

- <h4>Using Pie Chart For Share</h4>

```python
# Setting the figure size
plt.figure(figsize=(8, 8))

plt.pie(category_analysis['sales'], 
        labels=category_analysis['category'], 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=sns.color_palette('pastel'),
        explode=(0.05, 0, 0)) 
plt.title('Sales Market Share by Category', fontsize=14, fontweight='bold')

plt.savefig('category_share_pie.png')

plt.show()
```

<h3> WHICH ARE THE TOP 10 PRODUCTS BY SALES?</h3>

```python
#  Grpuping the Product coloumn
prod_sales = pd.DataFrame(df.groupby('product_name')['sales'].sum().reset_index())
```

<h3> WHICH ARE THE MOST SELLING PRODUCTS?</h3>

<h3> WHAT IS THE MOST PREFERRED SHIP MODE?

```python
# PREFERRED SHIP MODE using Bar chart
```

```python
# counting the ship mode types
ship_mode_counts = df['ship_mode'].value_counts()
```

```python
 # Setting the Figure size
plt.figure(figsize=(10, 6))

sns.barplot(x=ship_mode_counts.index, y=ship_mode_counts.values, palette='viridis')


plt.title('Preferred Ship Modes', fontsize=14, fontweight='bold')
plt.xlabel('Ship Mode', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.savefig('Preferred Ship Modes.png')

plt.show()
```
